```python
import pandas as pd
```

```python
df = pd.read_excel('JAWA-TIMUR-KOTA-SURABAYA-SEMUA-DATA-Layak-report.xlsx', names=['data'])
df = df.dropna().reset_index(drop=True)
# df = df.drop([0], axis=0)
df.iloc[:20]
```

```python
def split_messy_row(data_messy):
    row = []
    row_split = []
    is_dot = False
    for key in data_messy.split():
        if(is_dot):
            row.append(key)
            row_split.append(row)
            row = []
            is_dot = False
            continue
        if(key == ':'):
            is_dot = True
        row.append(key)

    row_split

    rows = []
    for r in row_split:
        rows.append(' '.join(r))
    return rows
```

```python
df.data
```

```python
row = []
dataset = []
is_nrow = False
for r in df.data:
    if(r == 'NO\nDATA\nFOTO'):
        continue
        
    if(isinstance(r, int)):
        dataset.append(row)
        row = []
        continue
    if(r.count('\t') > 1):
        for spr in split_messy_row(r):
            row.append(spr.split(':')[1].strip())
        continue
    r = r.replace("\t", "")
    r = r.split(':')[1].strip()
    row.append(r)
dataset.pop(0)
dataset
```

```python
df_output = pd.DataFrame(columns=('NIK Target Survey', 'Nama Lengkap', 'Propinsi, Kab/Kota', 'Kecamatan, Desa/Kel', 'Alamat',
                                  'Koordinat', 'Ada tempat berteduh tetap', 'Tinggal bersama anggota keluarga',
                                  'KK bekerja', 'Pernah khawatir /pernah tidak makan', 'Pengeluaran pangan > 70% pengeluaran',
                                  'Ada pengeluaran pakaian dalam setahun', 'Sebagian besar berlantai tanah',
                                  'Sebagian besar berdinding bambu / kawat / kayu', 'Memiliki jamban sendiri',
                                  'Penerangan dari listrik 450VA / bukan listrik', 'Pekerjaan Hasil Survey'))
for i in range(len(dataset)):
    df_output.loc[i] = dataset[i]

df_output
```

```python
df_output.to_csv('layak-jawa timur-surabaya.csv', index=False)
```
