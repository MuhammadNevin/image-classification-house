# Get Questionnaire Data

```python
def get_questionnaire(url):
    myFile=open(url,'r')
    soup = BeautifulSoup(myFile, "html.parser")
    
    # Get column name
    column = []
    for record in soup.find_all('table')[1:]:
        column = [tr.text.split(" : ")[0] for tr in record.find_all('tr')]
        break

    l = []
    for record in soup.find_all('table')[1:]:
        row = [tr.text.split(" : ")[1] for tr in record.find_all('tr')]
        l.append(row)

    return pd.DataFrame(l, columns=column)
```

# Get Image Data

```python
size = 1000

def download_image(url, start, label):
    myFile=open(url,'r')
    soup = BeautifulSoup(myFile, "html.parser")
    img = soup.find_all("img")
    file_name = start
    for i in img:
        image_link = i['src'].split('&')[0] + '&size=' + str(size)
        img_data = requests.get(image_link).content
        with open('img_1000/'+label+'/'+str(file_name)+'.jpg', 'wb') as handler:
            handler.write(img_data)
        file_name += 1
```
