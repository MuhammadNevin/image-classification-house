```python
class_names = ['layak', 'tidak-layak']
class_names_label = {class_name:i for i, class_name in enumerate(class_names)}

nb_classes = len(class_names)

IMAGE_SIZE = (150, 150)
```

```python
def load_data():
    """
        Load the data:
            - 14,034 images to train the network.
            - 3,000 images to evaluate how accurately the network learned to classify images.
    """
    
    datasets = ['../1_WrappingData/img/data_train']
    output = []
    
    # Iterate through training and test sets
    for dataset in datasets:
        
        images = []
        labels = []
        
        print("Loading {}".format(dataset))
        
        # Iterate through each folder corresponding to a category
        
        # natsort_file_names = natsorted(file_names)        
        for folder in os.listdir(dataset):
            label = class_names_label[folder]
            
            # Iterate through each image in our folder
            for file in tqdm(natsorted(os.listdir(os.path.join(dataset, folder)))):
                
                try:
                
                    # Get the path name of the image
                    img_path = os.path.join(os.path.join(dataset, folder), file)

                    # Open and resize the img
                    image = cv2.imread(img_path)
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #                 image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#                     image = cv2.resize(image, IMAGE_SIZE) 

                    # Append the image and its corresponding label to the output
                    images.append(image)
                    labels.append(label)
                    
                except:
                    print("error read image", file)
                    print(image)
                
#         images = np.array(images, dtype = 'float32')
        labels = np.array(labels, dtype = 'int32')   
        
        output.append((images, labels))

    return output
```

```python
[(train_images, train_labels)] = load_data()
```

```python
def get_data(data_detected):
    data = []
    object_detect = []
    flag = 0
    image_name = str()

    for i, val in data_detected[9:].iterrows():
        if('drive' in val[0]):
            image_name = val[0].split('/')[-1].split(':')[0]
            flag = 1
            continue
        if('Enter Image Path' in val[0]):
            data.append([image_name, object_detect])
            object_detect = []
            flag = 0
        if(flag):
            object_detect.append(val[0].split(':')[0])
            
    return pd.DataFrame(columns=['image_name', 'objects'], data=data)
```

```python
def display_image(class_names, images, labels, ind):
    """
        Display a random image from the images array and its correspond label from the labels array.
    """
    
    index = ind-1
    plt.figure()
    plt.imshow(images[index])
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title('Image #{} : '.format(ind) + class_names[labels[index]])
    plt.show()
```

```python
display_image(class_names, train_images, train_labels, 5)
```

```python
def show_image_by_object(class_names, images, labels, objects, num):
    file_list = list(df_tidak_layak_group_explode[df_tidak_layak_group_explode['objects'] == objects]['image_name'][:num])
    row = int((num-1)/4)+1
    
    fig = plt.figure(figsize=(15,5))
    fig.suptitle("Foto {num} teratas yang mengandung objek {objects}", fontsize=16)
    for i in range(num):
        plt.subplot(row,4,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        file = int(file_list[i].split('.')[0])
        
        plt.imshow(images[file-1], cmap=plt.cm.binary)
        
        plt.xlabel(class_names[labels[file-1]])
    plt.show()
```
