### Pembobotan kecamatan/desa

```python
layak_q[layak_q['Propinsi, Kab/Kota'] == values[0]]['Kecamatan, Desa/Kel'].value_counts()
```

```python
bobot_kec_des_all = []
df_bobot = pd.DataFrame({'kecamatan':[], 'bobot kecamatan':[], 'bobot final':[]})
for k in values:
    kec_name = layak_q[layak_q['Propinsi, Kab/Kota'] == k]['Kecamatan, Desa/Kel'].value_counts().keys().tolist()
    kec_count =  layak_q[layak_q['Propinsi, Kab/Kota'] == k]['Kecamatan, Desa/Kel'].value_counts().tolist()
    kec_bob = [c/len(df[df['Propinsi, Kab/Kota'] == k]) for c in kec_count]
    bobot_kotkab = kab_kot[kab_kot['Kota/Kabupate'] == k].bobot.iloc[0]
    bobot_final = [c/len(df[df['Propinsi, Kab/Kota'] == k]) * bobot_kotkab for c in kec_count]
    
    bobot_kec_des_all.append([kec_name, kec_bob, bobot_final])
    df_temp = pd.DataFrame({'kecamatan':kec_name, 'bobot kecamatan':kec_bob, 'bobot final':bobot_final})
    df_bobot = df_bobot.append(df_temp, ignore_index=True)
```

# Pengeluaran Daerah

```python
pk = pd.read_excel('persentase_konsumsi.xlsx')
pk.head()
```

```python
dict_persentase = {}
dict_rata = {}
dict_pengeluaran = {}
for i, r in pk.iterrows():
    dict_persentase[r.Kota] = r['Persentase Rata-rata Pengeluaran Makanan Per Kapita']
    dict_rata[r.Kota] = r['Rata-rata Pengeluaran Makanan Per Kapita']
    dict_pengeluaran[r.Kota] = r['Pengeluaran Per Kapita ']
```

```python
tidak_layak_q['Persentase Rata-rata Pengeluaran Makanan Per Kapita'] = tidak_layak_q['Propinsi, Kab/Kota'].map(dict_persentase)
tidak_layak_q['Rata-rata Pengeluaran Makanan Per Kapita'] = tidak_layak_q['Propinsi, Kab/Kota'].map(dict_rata)
tidak_layak_q['Pengeluaran Per Kapita'] = tidak_layak_q['Propinsi, Kab/Kota'].map(dict_pengeluaran)
tidak_layak_q.head()
```

```python
layak_q['Persentase Rata-rata Pengeluaran Makanan Per Kapita'] = layak_q['Propinsi, Kab/Kota'].map(dict_persentase)
layak_q['Rata-rata Pengeluaran Makanan Per Kapita'] = layak_q['Propinsi, Kab/Kota'].map(dict_rata)
layak_q['Pengeluaran Per Kapita'] = layak_q['Propinsi, Kab/Kota'].map(dict_pengeluaran)
layak_q.head()
```

```python
layak_q.to_csv('layak_bobot.csv', index=False)
tidak_layak_q.to_csv('tidak_layak_bobot.csv', index=False)
```
